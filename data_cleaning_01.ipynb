```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Datasets/customer_call_list.xlsx')
df
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.drop(columns='Not_Useful_Column', inplace=True)
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('._/')
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace(r'\D', '', regex=True)
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace(r'(\d{3})(\d{3})(\d{4})', r'\1-\2-\3', regex=True)
df
```

```python
df['Phone_Number'] = df['Phone_Number'].fillna('')
df
```

```python
df[['Street', 'State', 'ZIP']] = df['Address'].str.split(',', n=2, expand=True)
df
```

```python
df.drop(columns='Address', inplace=True)
df
```

```python
df = df[['CustomerID', 'First_Name', 'Last_Name', 'Phone_Number', 'Street', 'State', 'ZIP', 'Paying Customer', 'Do_Not_Contact']]
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df
```

```python
df = df.replace(['NaN', 'N/a', 'None', '', ' '], np.nan)
df
```

```python
df.fillna('', inplace=True)
df
```
